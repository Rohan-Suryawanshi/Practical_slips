```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

```python
data=[['eggs', 'milk','bread'], ['eggs', 'apple'], ['milk', 'bread'], ['apple', 
'milk'], ['milk', 'apple', 'bread']] 

data
```

```python
tr=TransactionEncoder()
new_frame=tr.fit(data).transform(data)

frame=pd.DataFrame(new_frame,columns=tr.columns_)
frame
```

```python
frequrnt_items=apriori(frame,min_support=0.1,use_colnames=True)
frequrnt_items
```

```python
asscoiction=association_rules(frequrnt_items)
asscoiction
```
